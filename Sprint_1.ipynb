```python
Contexto0 = '''
 Dados Reais sobre Energia Solar no Brasil (2024)
 Custo Médio de Instalação
Residencial (3 a 5 kWp): R$ 15.000 a R$ 35.000 (depende do tamanho e tecnologia dos painéis).

Comercial (10 a 50 kWp): R$ 50.000 a R$ 200.000.

Fator que influencia: Qualidade dos painéis (monocristalinos são mais eficientes, porém mais caros).

 Economia na Conta de Luz
Redução de 70% a 95% no valor da conta.

Exemplo: Uma casa que paga R$ 500/mês pode ter gastos reduzidos para R$ 100/mês ou menos.

Empresas com alto consumo (R$ 5.000/mês) podem economizar R$ 4.000/mês.

 Tempo de Retorno do Investimento (Payback)
Residencial: 4 a 7 anos (depende do consumo e tarifa local).

Comercial: 3 a 5 anos (devido a incentivos fiscais e maior escala).

 Vida Útil do Sistema
Painéis solares duram 25 a 30 anos.

Inversores precisam de troca após 10 a 15 anos.

 Manutenção
Limpeza dos painéis a cada 6 meses (se houver muita poeira).

Custo baixo: R$ 200 a R$ 500/ano.

 Financiamento e Incentivos
Bancos como Banco do Brasil, Caixa e Santander oferecem linhas de crédito com juros a partir de 0,8% ao mês.

Alguns estados têm isenção de ICMS para microgeração.

 Geração de Energia
1 kWp (quilowatt-pico) gera ~150 kWh/mês no Brasil (varia por região).

Exemplo: Um sistema de 5 kWp gera ~750 kWh/mês (suficiente para uma família de 4 pessoas).

 Documentação Necessária
Projeto elétrico aprovado.

Aprovação da distribuidora de energia (concessionária).

Laudo de conformidade (ART do engenheiro).



 Dados Técnicos e Mercadológicos Completos
1. Tipos de Sistemas Disponíveis:

On-grid (conectado à rede): 95% das instalações no Brasil

Off-grid (com baterias): para locais remotos sem acesso à rede

Híbrido: combina as duas tecnologias

2. Componentes do Sistema:

Painéis fotovoltaicos: eficiência média de 19-22% (monocristalinos)

Inversores: vida útil de 8-12 anos (string) ou 15+ anos (microinversores)

Estruturas de fixação: alumínio ou aço galvanizado

Sistemas de monitoramento: acompanhamento em tempo real via app

3. Potencial de Geração por Região:

Nordeste: 5,9 kWh/kWp/dia (melhor irradiação)

Sudeste: 4,7 kWh/kWp/dia

Sul: 4,2 kWh/kWp/dia

4. Legislação e Regulamentação:

Resolução Normativa 1.000/2021 da ANEEL

Lei 14.300/2022 (Marco Legal da GD)

Regras de compensação de créditos energéticos

Prazos de análise pelas distribuidoras: até 34 dias úteis

5. Aspectos Financeiros Detalhados:

Taxa interna de retorno (TIR) média: 18-30% ao ano

Valorização do imóvel: 3-6% segundo estudos do SEBRAE

Linhas de financiamento específicas:

BNDES Finame - juros a partir de 0,9% a.m.

Cartão BNDES - até R$ 500 mil

BB Energia Solar - até 100% do projeto

6. Manutenção e Garantias:

Garantia dos painéis: 10-12 anos contra defeitos, 25-30 anos de performance

Limpeza recomendada: 2 vezes/ano em áreas urbanas

Custo médio de manutenção anual: 0,5% do valor do sistema

7. Processo de Instalação Passo a Passo:

Análise de viabilidade técnica

Projeto elétrico detalhado

Homologação na concessionária

Instalação física (1-5 dias)

Vistoria e ligação do sistema

Ativação do monitoramento

8. Mitos e Verdades:

Funciona em dias nublados? (Sim, com 10-30% da capacidade)

Precisa de baterias? (Não para sistemas on-grid)

Valoriza mesmo o imóvel? (Sim, comprovado por estudos)

Pode faltar energia? (Sim, desliga durante blecaute por segurança)

9. Comparativo de Tecnologias:

Monocristalino: + eficiente (22%), + caro

Policristalino: 18% eficiência, custo médio

Filme fino: 10% eficiência, mas flexível

10. Perguntas Frequentes Avançadas:

Como dimensionar corretamente o sistema?

O que considerar no contrato com a instaladora?

Como funciona a compensação de créditos?

Quais os riscos de uma instalação mal feita?

Como calcular o ROI para empresas?

11. Tendências do Mercado:

Crescimento médio de 40% ao ano no Brasil

Novas tecnologias: painéis bifaciais, telhas solares

Armazenamento: queda de 80% no preço de baterias desde 2015

12. Casos Reais de Sucesso:

Residência em SP: investimento de R$ 25k, payback em 4 anos

Comércio em MG: redução de R$ 3.500 para R$ 400 na conta

Indústria no PR: sistema de 1MW, retorno em 3,2 anos

13. Alertas Importantes:

Cuidado com orçamentos muito abaixo do mercado

Verificar certificações dos equipamentos

Exigir ART (Anotação de Responsabilidade Técnica)

Checar referências da empresa instaladora

14. Dicas de Otimização:

Orientação ideal dos painéis: Norte verdadeiro

Inclinação de acordo com a latitude local

Sombreamento zero entre 9h e 15h

Uso de optimizers em caso de sombras parciais

15. Suporte Pós-Venda:

Monitoramento remoto do sistema

Atualizações de firmware

Assistência técnica especializada

Relatórios mensais de geração

Orientações para o Assistente:

Sempre contextualize as respostas com dados reais

Forneça exemplos práticos quando possível

Indique quando é necessário consultar um especialista presencial

Atualize constantemente as informações sobre incentivos fiscais

Destaque os benefícios ambientais (cada kWp evita ~30kg CO2/mês)


'''
```

```python
msgs = [{'role': 'system', 'content': 'Você é um especialista em energia solar, ajudando usuários a entenderem os benefícios, custos e processos de instalação de sistemas fotovoltaicos em residências ou empresas. Responda de forma clara e prática, com dados reais e atualizados do mercado brasileiro (2024).' + Contexto0}]

while True:
    pergunta = input()


    msgs.append({'role': 'user', 'content': pergunta})
    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=msgs
    )
    resposta = completion.choices[0].message.content
    print(resposta)
    msgs.append({'role': 'assistant', 'content': resposta})
```
